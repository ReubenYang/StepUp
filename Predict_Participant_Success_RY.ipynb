```python
#Vriables used in model
var_list = ['Criminal_record__c', 'Drivers_License__c', 'Substance_Abuse__c', 'Homeless__c']
X = Data.ix[:, var_list]
Y = Data.Job_ready__c
X.head()
X.describe()
#Y.describe()
```

```python
#Train-test split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score, train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=4231)
#print(X_train.head(), y_train.head())
```

```python
#Cross Validation
from sklearn.metrics import accuracy_score

def cv_score(clf, x, y):
    result = 0
    nfold = 5
    i = 0
    for train, test in KFold(nfold).split(x): # split data into train/test groups, 5 times
        i += 1
        #print(x.iloc[train,:].head())
        #print(y.iloc[train].head())
        clf.fit(x.iloc[train,:], y.iloc[train].values.ravel()) # fit
        y_pred = clf.predict(x.iloc[test,:]) #predict
        print("Accuracy Score:", accuracy_score(y.iloc[test].values.ravel(), y_pred), "for Fold ", i)
        result += accuracy_score(y.iloc[test].values.ravel(), y_pred) # evaluate accury score on held-out data
        
    return result / nfold # average
```

# Logistic Regression 

```python
#Logistic Regression (tuning regularization parameter)
Cs = [0.01, 0.1, 1, 10, 100]
max_score = 0
C_max = 0

for c in Cs:
    print("C = ", c)
    clf_c = LogisticRegression(C=c)
    score_c = cv_score(clf_c, X_train, y_train)
    print("Average Accuracy Score:", score_c)
    print("---------------------------------------------------")
    if score_c > max_score:
        max_score = score_c
        C_max = c
        
print("Best C parameter:", C_max, ", ", "Maximum Accuracy Score:", max_score)
```

```python
#Use the best C on test data
clf_t = LogisticRegression(C=C_max)
clf_t.fit(X_train, y_train) # fit
y_pred_test = clf_t.predict(X_test) #predict
print("Accuracy Score on Test Data:", accuracy_score(y_test, y_pred_test))
```

# SVM

```python
#SVM with Linear Kernel
from sklearn import svm
from sklearn.model_selection import GridSearchCV
parameters = [{'C': [1, 10, 100, 1000], 'kernel': ['linear']}]

Cs = [1, 10, 100, 1000]

max_score = 0
C_max = 0

for c in Cs:
    print("C = ", c)
    clf_c = svm.SVC(C=c, kernel='linear')
    score_c = cv_score(clf_c, X_train, y_train)
    print("Average Accuracy Score:", score_c)
    print("---------------------------------------------------")
    if score_c > max_score:
        max_score = score_c
        C_max = c
        
print("Best C parameter:", C_max, ", ", "Maximum Accuracy Score:", max_score)
```

```python
#Use the best C on test data
clf_t = svm.SVC(C=C_max, kernel='linear')
clf_t.fit(X_train, y_train) # fit
y_pred_test = clf_t.predict(X_test) #predict
print("Accuracy Score on Test Data:", accuracy_score(y_test, y_pred_test))
```

```python
#SVM with Radial Basis Function 

Cs = [1, 10, 100, 1000]
Gammas = [0.001, 0.0001]

max_score = 0
C_max = 0
g_max = 0 

for c in Cs:
    for g in Gammas:
        print("C = ", c, ' Gamma = ', g)
        clf_c = svm.SVC(C=c, kernel='rbf', gamma=g)
        score_c_g = cv_score(clf_c, X_train, y_train)
        print("Average Accuracy Score:", score_c_g)
        print("---------------------------------------------------")
        if score_c_g > max_score:
            max_score = score_c_g
            C_max = c
            g_max = g
        
print("Best C parameter:", C_max, ", ", "Best Gamma Parameter:", g_max, ", " "Maximum Accuracy Score:", max_score)
```

```python
#Use the best C and Gamma on test data
clf_x = svm.SVC(C=C_max, kernel='rbf', gamma=g_max)
clf_x.fit(X_train, y_train) # fit
y_pred_test = clf_x.predict(X_test) #predict
print("Accuracy Score on Test Data:", accuracy_score(y_test, y_pred_test))
```
